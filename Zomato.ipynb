```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names=final_df['Country'].value_counts().index
country_names
```

```python
country_values=final_df['Country'].value_counts().values
```

```python
#pie chart--top 3 countries
plt.pie(country_values[:3],labels=country_names[:3],autopct='%0.1f%%')
plt.show()
```

**Observations/Conclusions:**
    Zomoto maximum records or transactions are from India
    After that usa and then uk

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

# **Conclusions**

1.when rating is between 4.5 -4.9 ----> Excellemnt
2.when rating is between 4.0to4.4  ----> very good
3.when rating is between    3.5 to 3.9----> good
4.when rating is between   3.0 to 3.4----->average
5.when rating is between  2.5 to 2.9 ----->average
6.when rating is between   2.0 to 2.4 --->poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
#plt.figure(figsize=(12, 6))  # Adjust the width and height as needed
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['blue','white','red','orange','yellow','green','green'])
```

Obervation:
1. Not rated count is very high
2. Maximum number of rating are b/w 2.5 to 3.4


```python
#Count plot
sns.countplot(x='Rating color',data=ratings,palette=['blue','white','red','orange','yellow','green','green'])
```
